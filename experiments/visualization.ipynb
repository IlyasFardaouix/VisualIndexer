# Data Visualization

Generate various plots to understand the dataset.

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn-v0_8-whitegrid')
%matplotlib inline
```

## Load Data

```
df = pd.read_csv('data.csv')
df.head()
```

## Pair Plot

```
numeric_cols = df.select_dtypes(include='number').columns[:5]  # first 5 numeric
sns.pairplot(df[numeric_cols])
plt.suptitle('Pairwise Relationships', y=1.02)
plt.show()
```

## Box Plots

```
fig, axes = plt.subplots(1, min(4, len(numeric_cols)), figsize=(16, 4))
if not hasattr(axes, '__len__'):
    axes = [axes]
for ax, col in zip(axes, numeric_cols):
    sns.boxplot(y=df[col], ax=ax)
    ax.set_title(col)
plt.tight_layout()
plt.show()
```

## Value Counts for Categorical Columns

```
cat_cols = df.select_dtypes(include='object').columns
for col in cat_cols[:3]:  # first 3 categorical
    plt.figure(figsize=(8, 4))
    df[col].value_counts().head(15).plot(kind='bar')
    plt.title(f'Top values: {col}')
    plt.tight_layout()
    plt.show()
```
